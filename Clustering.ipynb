```python
df_scaled_data=df.drop('ID#',axis=1)
df_scaled_data
```

```python
# Standardizing the data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_df=scaler.fit_transform(df_scaled_data)
scaled_df
```

```python
from sklearn.cluster import KMeans
# Standardize the data
scaler = StandardScaler()
x = scaler.fit_transform(df_scaled_data)


# Apply K-Means with the optimal number of clusters (set manually or from Elbow Method)
optimal_k = 5  # Adjust based on the Elbow curve analysis
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
y_means = kmeans.fit_predict(x)

# Scatter plot for cluster visualization
plt.figure(figsize=(8, 6))

# Assigning colors for up to 8 clusters (modify as needed)
colors = ['pink', 'yellow', 'cyan', 'magenta', 'orange', 'blue', 'red', 'black', 'violet']

for i in range(optimal_k):
    plt.scatter(x[y_means == i, 0], x[y_means == i, 1], s=50, c=colors[i], label=f'Cluster {i}')

# Scatter plot for cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            s=200, c='green', marker='X', label='Centroids')

plt.xlabel("Balance (Standardized)")
plt.ylabel("Bonus Miles (Standardized)")
plt.title("Clusters of EastWestAirlines Customers")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
# Finding optimal K using the Elbow Method
inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_df)
    inertia.append(kmeans.inertia_)
```

```python
# Plot Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o', linestyle='--')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal K in K-Means")
plt.show()
```

```python
# Apply K-Means with chosen K
optimal_k = 4  # Adjust based on Elbow method
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df_scaled_data["KMeans_Cluster"] = kmeans.fit_predict(scaled_df)
df_scaled_data["KMeans_Cluster"]
```

```python
# Apply Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=optimal_k, linkage="ward")
df_scaled_data["Hierarchical_Cluster"] = hierarchical.fit_predict(scaled_df)
df_scaled_data["Hierarchical_Cluster"]
```

```python
# Apply DBSCAN (Tune eps and min_samples as needed)
dbscan = DBSCAN(eps=1.5, min_samples=5)
df_scaled_data["DBSCAN_Cluster"] = dbscan.fit_predict(scaled_df)
df_scaled_data["DBSCAN_Cluster"]
```

```python
# Plot Dendrogram
from scipy.cluster.hierarchy import dendrogram, linkage
plt.figure(figsize=(10, 5))
linkage_matrix = linkage(scaled_df, method="ward")
dendrogram(linkage_matrix)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
# Display results
print(df_scaled_data.head())
```

```python
# Cluster Analysis and Interpretation
# K-Means Clustering
# Attach cluster labels to original data
df["KMeans_Cluster"] = y_means

# Group by cluster and view the mean of each feature
cluster_summary = df.groupby("KMeans_Cluster").mean()
print(cluster_summary)
```

```python
# Hierarchical Clustering
# (This was already done earlier using the dendrogram)
# You can also extract clusters with:
from scipy.cluster.hierarchy import fcluster
labels = fcluster(linkage_matrix, t=5, criterion='maxclust')
df["Hierarchical_Cluster"] = labels
df.groupby("Hierarchical_Cluster").mean()
```

```python
# DBSCAN
df["DBSCAN_Cluster"] = dbscan.labels_
df["DBSCAN_Cluster"].value_counts()  # Check cluster distribution
```

```python
# Visualization
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Balance"], y=df["Bonus_miles"], hue=df["KMeans_Cluster"], palette="Set1", s=60)
plt.title("K-Means Clustering")
plt.xlabel("Balance")
plt.ylabel("Bonus Miles")
plt.legend(title="Cluster")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Balance"], y=df["Bonus_miles"], hue=df["Hierarchical_Cluster"], palette="Set2", s=60)
plt.title("Hierarchical Clustering")
plt.xlabel("Balance")
plt.ylabel("Bonus Miles")
plt.legend(title="Cluster")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Balance"], y=df["Bonus_miles"], hue=df["DBSCAN_Cluster"], palette="tab10", s=60)
plt.title("DBSCAN Clustering")
plt.xlabel("Balance")
plt.ylabel("Bonus Miles")
plt.legend(title="Cluster (DBSCAN)")
plt.show()
```

```python
# Evaluation and Performance Metrics:
from sklearn.metrics import silhouette_score

# Evaluate K-Means Clustering
kmeans_score = silhouette_score(x, df["KMeans_Cluster"])
print(f"Silhouette Score for K-Means: {kmeans_score:.3f}")

# Evaluate DBSCAN (excluding noise points labeled -1)
mask = df["DBSCAN_Cluster"] != -1  # Filter out noise
if df["DBSCAN_Cluster"].nunique() > 1:
    dbscan_score = silhouette_score(x[mask], df["DBSCAN_Cluster"][mask])
    print(f"Silhouette Score for DBSCAN (excluding noise): {dbscan_score:.3f}")
else:
    print("DBSCAN produced only one cluster or only noise â€” Silhouette Score not applicable.")
```

Conclusion:
* The clustering analysis using K-Means, Hierarchical Clustering, and DBSCAN demonstrated that each algorithm has unique strengths. K-Means performed best when clusters were well-separated and spherical, giving the highest silhouette score. Hierarchical clustering provided an intuitive visualization through dendrograms, helping identify natural groupings, while DBSCAN effectively detected noise and handled irregular cluster shapes but required careful parameter tuning. Overall, the most suitable method depends on the datasetâ€™s distribution, presence of noise, and the need for interpretability
