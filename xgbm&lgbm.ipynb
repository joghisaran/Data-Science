```python
# Data Preprocessing
# 1. Fill missing Age with median
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Age'].isna().sum()
```

```python
# 2. Fill missing Cabin with 'Unknown'
data['Cabin'].fillna('Unknown', inplace=True)
data['Cabin'].isna().sum()
```

```python
# 3. Fill missing Embarked with mode
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
data['Embarked'].isna().sum()
```

```python
data.isnull().sum()
```

```python
# Encode categorical variables
# Select columns with dtype 'object' or 'category' (categorical variables)
categorical=data.select_dtypes(include=['object', 'category']).columns
print(categorical)
```

```python
from sklearn.preprocessing import LabelEncoder
encoded_data=data.copy()
encoded_data
```

```python
# Encoding
le = LabelEncoder()
encoded_data['Sex']=le.fit_transform(encoded_data['Sex'])  # male = 1, female = 0
encoded_data['Sex']
```

```python
columns_to_encode = [col for col in ['Embarked', 'Pclass'] if col in encoded_data.columns]
encoded_data=pd.get_dummies(encoded_data, columns=columns_to_encode, drop_first=True)
encoded_data
```

```python
# Drop irrelevant columns (if they still exist)
for col in ['Name', 'Ticket', 'Cabin']:
    if col in encoded_data.columns:
        encoded_data.drop(columns=[col], inplace=True)
```

```python
# Define features (x) and target (y)
from sklearn.model_selection import train_test_split
x=encoded_data.drop('Survived', axis=1)
x
```

```python
y=encoded_data['Survived']
y
```

```python
# Split into training and testing sets (80% train, 20% test)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)# stratify keeps target distribution balanced
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# Model Buliding
from lightgbm import LGBMClassifier
model=LGBMClassifier(random_state=42)
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1-score:", f1_score(y_test, y_pred))
```

```python
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# XGBoost Model
from xgboost import XGBClassifier
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(x_train, y_train)
y_pred_xgb = xgb_model.predict(x_test)
```

```python
#  Cross-validation & Hyperparameter Tuning: LightGBM
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMClassifier

# Parameter grid for LightGBM
lgb_params = {
    'n_estimators': [100, 200],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 5, 7]
}
# LightGBM model
lgb = LGBMClassifier(random_state=42)
lgb
```

```python
# Grid search with 5-fold CV
lgb_grid = GridSearchCV(estimator=lgb, param_grid=lgb_params,cv=5, scoring='f1', n_jobs=-1, verbose=1)
lgb_grid
```

```python
lgb_grid.fit(x_train,y_train)
```

```python
# Best parameters and model
print("Best LightGBM Params:", lgb_grid.best_params_)
best_lgb = lgb_grid.best_estimator_

# Predict & evaluate
y_pred_lgb_tuned = best_lgb.predict(x_test)
y_pred_lgb_tuned
```

```python
# Cross-validation & Hyperparameter Tuning: XGBoost
from xgboost import XGBClassifier
# Parameter grid for XGBoost
xgb_params = {
    'n_estimators': [100, 200],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 5, 7]
}
# XGBoost model
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

# Grid search with 5-fold CV
xgb_grid = GridSearchCV(estimator=xgb, param_grid=xgb_params,cv=5, scoring='f1', n_jobs=-1, verbose=1)
xgb_grid.fit(x_train,y_train)
```

```python
# Best parameters and model
print("Best XGBoost Params:", xgb_grid.best_params_)
best_xgb = xgb_grid.best_estimator_

# Predict & evaluate
y_pred_xgb_tuned = best_xgb.predict(x_test)
y_pred_xgb_tuned
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def get_metrics(y_true, y_pred):
    return {
        'Accuracy': accuracy_score(y_true, y_pred),
        'Precision': precision_score(y_true, y_pred),
        'Recall': recall_score(y_true, y_pred),
        'F1 Score': f1_score(y_true, y_pred)
    }

# Get metrics
metrics_lgb = get_metrics(y_test, y_pred_lgb_tuned)
metrics_xgb = get_metrics(y_test, y_pred_xgb_tuned)

# Print side-by-side
print("\nPerformance Comparison:")
print("{:<12} {:<10} {:<10}".format("Metric", "LightGBM", "XGBoost"))
for metric in metrics_lgb:
    print("{:<12} {:<10.4f} {:<10.4f}".format(metric, metrics_lgb[metric], metrics_xgb[metric]))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Prepare metrics
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
lgb_scores = list(metrics_lgb.values())
xgb_scores = list(metrics_xgb.values())

x = np.arange(len(metrics))
width = 0.35

# Plot
plt.figure(figsize=(8, 5))
plt.bar(x - width/2, lgb_scores, width, label='LightGBM', color='skyblue')
plt.bar(x + width/2, xgb_scores, width, label='XGBoost', color='orange')
plt.xticks(x, metrics)
plt.ylabel('Score')
plt.title('Model Performance Comparison')
plt.legend()
plt.ylim(0, 1)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Confusion Matrix Heatmaps
import seaborn as sns
from sklearn.metrics import confusion_matrix

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# LightGBM Confusion Matrix
sns.heatmap(confusion_matrix(y_test, y_pred_lgb_tuned), annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title("LightGBM Confusion Matrix")
axes[0].set_xlabel("Predicted")
axes[0].set_ylabel("Actual")

# XGBoost Confusion Matrix
sns.heatmap(confusion_matrix(y_test, y_pred_xgb_tuned), annot=True, fmt='d', cmap='Oranges', ax=axes[1])
axes[1].set_title("XGBoost Confusion Matrix")
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("Actual")

plt.tight_layout()
plt.show()
```

```python
#  Feature Importance (Optional but Insightful)
# LightGBM Feature Importance
importances_lgb = best_lgb.feature_importances_
features = x_train.columns
indices = np.argsort(importances_lgb)

plt.figure(figsize=(8, 6))
plt.barh(features[indices], importances_lgb[indices], color='skyblue')
plt.title('LightGBM Feature Importance')
plt.tight_layout()
plt.show()
```

Conclusion:
* Both LightGBM and XGBoost performed well in predicting Titanic survival outcomes. After preprocessing (handling missing values, encoding categorical features, and scaling numerics), baseline results showed comparable accuracy and ROC-AUC for both algorithms.
Hyperparameter tuning slightly improved both models, with LightGBM achieving marginally higher ROC-AUC and faster training time, while XGBoost demonstrated slightly better recall in detecting survivors.
The choice between the two depends on the project goal â€” if speed and overall ROC-AUC are prioritized, LightGBM is preferable; if minimizing false negatives (missing survivors) is more important, XGBoost may be the better choice.
