```python
 #Load dataset
df = pd.read_csv("Cardiotocographic.csv")
```

```python
#Handle missing values
df.isnull().sum()  # Check for missing values
```

```python
#Calculating the percentage of missing values in each column
df.isnull().mean()*100
```

```python
df.fillna(df.median(), inplace=True)  # Impute if needed
```

```python
#Fix_datatypes
```

```python
# Example if a numeric column is stored as string
df["LB"] = pd.to_numeric(df["LB"], errors="coerce")
```

```python
#outlier_detection
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```
